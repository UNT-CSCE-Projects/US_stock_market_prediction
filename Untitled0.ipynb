```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your time series data into a pandas DataFrame
# For example, if your data is in a CSV file, you can use pd.read_csv('your_data.csv')

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(data['date_id'],data['target'], label='Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data')

# Calculate rolling statistics (e.g., moving average)
rolling_mean = data['target'].rolling(window=12).mean()
rolling_std = data['target'].rolling(window=12).std()

# Plot rolling statistics
plt.plot(data['date_id'], rolling_mean, label='Rolling Mean')
plt.plot(data['date_id'], rolling_std, label='Rolling Std')

plt.legend()
plt.show()

# Perform Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

result = adfuller(data['target'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])

if result[1] <= 0.05:
    print('The time series is stationary.')
else:
    print('The time series is not stationary.')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.stattools import adfuller

# Load your time series data into a pandas DataFrame
# Replace 'date_column' and 'value_column' with your own column names
# For example, if your data is in a CSV file, you can use pd.read_csv('your_data.csv')

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(data['time_id'],data['target'], label='Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data')

# Check for stationarity using ADF test
result = adfuller(data['target'])
if result[1] <= 0.05:
    print('The time series is stationary.')
else:
    print('The time series is not stationary. Please make it stationary before fitting an ARIMA model.')
    # You can use differencing to make it stationary.

# Fit an ARIMA model
# The order of the ARIMA model is represented as (p, d, q), where:
# p: The order of the autoregressive component.
# d: The order of differencing required to make the series stationary.
# q: The order of the moving average component.

# Example: Fit an ARIMA(1, 1, 1) model
p, d, q = 1, 1, 1
model = ARIMA(data['target'], order=(p, d, q))
results = model.fit(disp=-1)

# Print model summary
print(results.summary())

# Plot the fitted values
plt.figure(figsize=(12, 6))
plt.plot(data['time_id'], data['target'], label='Original Data')
plt.plot(data['time_id'], results.fittedvalues, color='red', label='Fitted Values')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('ARIMA Fit')
plt.legend()
plt.show()
```
